```python
import pandas as pd

# Define list of file paths
bts_files = ['JAN-2023.csv', 'FEB-2023.csv', 'MARCH-2023.csv','APRIL-2023.csv','MAY-2023.csv','JUNE-2023.csv']

# Concatenate multiple CSV files into a single DataFrame
flight_data_set = pd.concat([pd.read_csv(f) for f in bts_files])

# Display the first few rows of the concatenated DataFrame
flight_data_set.head()
```

```python
flight_data_set_filtered = flight_data_set[(flight_data_set['ORIGIN'].isin(['ORD', 'JFK', 'MCO'])) 
                                                & (flight_data_set['DEST'] == 'SYR') 
                                                    & (flight_data_set['MKT_UNIQUE_CARRIER'].isin(['AA','B6','DL','UA','WN']))]

print(set(flight_data_set_filtered['ORIGIN']))
print(set(flight_data_set_filtered['DEST']))
print(set(flight_data_set_filtered['MKT_UNIQUE_CARRIER']))
```

```python
# Define the file path where you want to save the filtered data
output_file_path = 'filtered_flight_data.csv'

# Save the filtered data to a CSV file
flight_data_set_filtered.to_csv(output_file_path, index=False)

print("Filtered data has been saved to:", output_file_path)
```
